Agora vamos realmente preencher o campo do dataset

```python
#Baseado na crosstab Title vs Age
titleXages_dict = { 'Capt'     : 'Elderly',
                    'Col'      : 'Adult',
                    'Countess' : 'Adult',
                    'Don'      : 'Adult',
                    'Dona'     : 'Adult', #SÃ³ encontrado em test
                    'Dr'       : 'Adult',
                    'Jonkheer' : 'Adult',
                    'Lady'     : 'Adult',
                    'Major'    : 'Adult',
                    'Master'   : 'Child',
                    'Miss'     : 'Adult', #Aqui temos trabalho ainda por fazer
                    'Mlle'     : 'Adult',
                    'Mme'      : 'Adult',
                    'Mr'       : 'Adult', #Aqui temos trabalho ainda por fazer, porÃ©m Ã© menos importante que Miss
                    'Mrs'      : 'Adult', #Aqui podemos realmente ignorar o trabalho por fazer
                    'Ms'       : 'Adult',
                    'Rev'      : 'Adult',
                    'Sir'      : 'Adult'}

#Para toda chave no Dict de Pronomes de Tratamento, se a Idade foi NaN sete a
#categoria da idade para o definidio no Dict
for key in titleXages_dict:
    titanic_train.loc[(titanic_train['Age'].isna()) & (titanic_train['Title'] == key), 'AgeCat'  ] = titleXages_dict[key]
    titanic_test.loc[(titanic_test['Age'].isna()) & (titanic_test['Title'] == key), 'AgeCat'  ] = titleXages_dict[key]

#Confirmamos que todas as idades foram preenchidas
print(titanic_train.isnull().sum())
print(titanic_test.isnull().sum())
```

Semelhante com o que fizemos com a variÃ¡vel contÃ­nua de Idade, vamos criar categorias para Fare (Valor da passagem).

ATENÃAO - O dataset tem algumas inconsistÃªncias em relaÃ§Ã£o ao valor da passagem que precisam de tratamento. Aqui ignoramos por questÃµes didÃ¡ticas.

```python
#Continuous variable bins; qcut vs cut: https://stackoverflow.com/questions/30211923/what-is-the-difference-between-pandas-qcut-and-pandas-cut
#Fare Bins/Buckets using qcut or frequency bins: https://pandas.pydata.org/pandas-docs/stable/generated/pandas.qcut.html
titanic_train['FareBin'] = pd.qcut(titanic_train['Fare'], 4)
titanic_test['FareBin'] = pd.qcut(titanic_test['Fare'], 4)
```

```python

#handmade data model using brain power (and Microsoft Excel Pivot Tables for quick calculations)
def mytree(df):

    #initialize table to store predictions
    Model = pd.DataFrame(data = {'Predict':[]})
    male_title = ['Master'] #survived titles

    for index, row in df.iterrows():

        #Question 1: Were you on the Titanic; majority died
        Model.loc[index, 'Predict'] = 0

        #Question 2: Are you female; majority survived
        if (df.loc[index, 'Sex'] == 'female'):
                  Model.loc[index, 'Predict'] = 1

        #Question 3A Female - Class and Question 4 Embarked gain minimum information

        #Question 5B Female - FareBin; set anything less than .5 in female node decision tree back to 0
        if ((df.loc[index, 'Sex'] == 'female') &
            (df.loc[index, 'Pclass'] == 3) &
            (df.loc[index, 'Embarked'] == 'S')  &
            (df.loc[index, 'Fare'] > 8)

           ):
                  Model.loc[index, 'Predict'] = 0

        #Question 3B Male: Title; set anything greater than .5 to 1 for majority survived
        if ((df.loc[index, 'Sex'] == 'male') &
            (df.loc[index, 'Title'] in male_title)
            ):
            Model.loc[index, 'Predict'] = 1


    return Model


#model data
Tree_Predict = mytree(titanic_train)
print('Decision Tree Model Accuracy/Precision Score: {:.2f}%\n'.format(metrics.accuracy_score(titanic_train['Survived'], Tree_Predict)*100))
```
