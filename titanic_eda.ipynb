```python
# Estratégia de categorias

def categorize_age(age: int):
  if age <= 0:
    return "no age"
  if age <= 10:
    return "child"
  if age <= 20:
    return "teen"
  if age <= 30:
    return "young adult"
  if age <= 50:
    return "adult"
  return "senior"

titanic_train['age'].fillna(-1, inplace=True)
titanic_train['age_category'] = titanic_train['age'].apply(categorize_age)
titanic_train['age_category'].value_counts()
```

Para classificadores, é uma boa ideia transformar dados contínuos em dados categóricos
(como já foi feito com idade e será feito a seguir com o preço do bilhete)

```python
def categorize_fare(fare: int):
  if fare <= 0:
    return "no payment"
  if fare <= 10:
    return "low payer (<10)"
  if fare <= 35:
    return "average payer (10:35)"
  if fare <= 135:
    return "high payer (35:135)"
  return "highest payer (>135)"

titanic_train['fare'].fillna(-1, inplace=True)
titanic_train['fare_category'] = titanic_train['fare'].apply(categorize_fare)
titanic_train['fare_category'].value_counts()
```

Podemos remover também os poucos NAN que sobraram.

```python
titanic_train = titanic_train.dropna()


print("Lista a quantidade de nulls por coluna")
print(titanic_train.isnull().sum())
print("Linhas e colunas do Dataset")
print(titanic_train.shape)
```

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()

titanic_train['Sex_Code'] = label.fit_transform(titanic_train['sex'])
titanic_train['Embarked_Code'] = label.fit_transform(titanic_train['embarked'])
titanic_train['age_code'] = label.fit_transform(titanic_train['age_category'])
titanic_train['fare_code'] = label.fit_transform(titanic_train['fare_category'])

titanic_train.head(5)
```

```python
from operator import xor
#Common Model Algorithms

from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split

from xgboost import XGBClassifier

#from sklearn import  tree
#model = tree.DecisionTreeClassifier()

import math

from sklearn import metrics

def train_model(df, columns: list, label: list):
    #modelo aprenda um f(x) = y
    train1_x, test1_x, train1_y, test1_y = train_test_split(df[columns], df[label], stratify=df[label], test_size=0.25, random_state = 42)

    print("dataset shapes:")
    print(train1_x.shape)
    print(test1_x.shape)
    print("")

    model = XGBClassifier()

    model.fit(train1_x, train1_y)

    pred_y = model.predict(test1_x)

    # print(pred_y)
    # print(test1_y)

    cf_matrix_TREE = confusion_matrix(test1_y, pred_y)

    print("Confusion Matrix:")
    print(cf_matrix_TREE)
    print("")


    TN, FP, FN, TP = cf_matrix_TREE.ravel()

    print("TN, FP, FN, TP")
    print(f"{TN}, {FP}, {FN}, {TP}")
    print("")

    #Existem métodos prontos para fazer esses cálculos, prefira sempre o uso de métodos prontos

    acuracia = (TP+TN) / (TN + FP + FN + TP)
    print(acuracia)
    sensibilidade = TP / (TP + FN)
    print(sensibilidade)
    especificidade = TN / (TN + FP)
    print(especificidade)
    precisao = TP / (TP + FP)
    print(precisao)

    phi = ( (TP*TN) - (FP*FN) )  / math.sqrt( (TP + FP) * (TP + FN) * (TN + FP) * (TN + FN) )
    print(phi)

    fpr, tpr, thresholds = metrics.roc_curve(test1_y, pred_y)
    auc = metrics.auc(fpr, tpr)
    print(auc) #ROC
```

Testing with continuous data

```python
x = ['pclass', 'sibsp', 'parch', 'fare', 'age', 'Sex_Code', 'Embarked_Code']
y = ['survived']

train_model(titanic_train, x, y)
```

Testing with categorical data

```python
x = ['pclass', 'sibsp', 'parch', 'Sex_Code', 'Embarked_Code', 'age_code', 'fare_code']
y = ['survived']

train_model(titanic_train, x, y)
```
